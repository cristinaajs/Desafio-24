```python
df = pd.read_csv('PDA_Lista_Instituicoes_Ensino_Superior_do_Brasil_EMEC.csv')
print(df.head())
```

```python
df.columns
```

```python
instituicoes_cidade = df['MUNICIPIO'].value_counts()
print(instituicoes_cidade)
```

```python
instituicoes_tipo = df['CATEGORIA_DA_IES'].value_counts()
print(instituicoes_tipo)
```

```python
instituicoes_estado = df['UF'].value_counts()
print(instituicoes_estado)
```

```python
instituicoes_organizacao = df['ORGANIZACAO_ACADEMICA'].value_counts()
print(instituicoes_organizacao)
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
instituicoes_estado.plot(kind='bar')
plt.ylabel('Quantidade de Universidades')
plt.xlabel('Estados')
plt.title('Universidades por Estado')
plt.show()
```

```python
instituicoes_estado.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('Quantidade de Universidades')
plt.xlabel('Estados')
plt.title('Universidades por Estado')
plt.show()
```

```python
instituicoes_cidade.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('Quantidade de Universidades')
plt.xlabel('Cidade')
plt.title('Universidades por Cidade')
plt.show()
```

```python
instituicoes_tipo.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('Tipo')
plt.xlabel('Tipo')
plt.title('Tipo de Universidades')
plt.show()
```

```python
instituicoes_organizacao.plot(kind='bar')
plt.ylabel('Tipo')
plt.xlabel('Tipo')
plt.title('Tipo de Universidades')
plt.show()
```
